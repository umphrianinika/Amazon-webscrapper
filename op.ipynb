```python
from bs4 import BeautifulSoup
import requests
import smtplib
import time
import datetime
```

```python
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36",
           "Accept-Language": "en-US,en;q=0.9"}
URL= 'https://www.amazon.de/-/en/HUGO-BOSS-Mens-T-Shirt-Pack/dp/B09QMBZB1K/ref=sr_1_8?s=photo&sr=1-8'

page = requests.get(URL, headers=headers)

soup1 = BeautifulSoup(page.content, "lxml")

title_tag = soup1.find(id='title')
if title_tag is None:
    print("title not found")
else:
    title = title_tag.get_text(strip=True)
price_ranges = soup1.select_one("td.a-span12").get_text(strip=True)
price = price_ranges.split('€')
prices = price[1]+ "-" +price[3]



```

```python
import datetime

today = datetime.date.today()
print(today)
```

```python
import csv

header = ['Title', 'Price', 'Date']
data = [title, prices, today]

with open('AmazonWebScraper.csv','w',newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerow(data)


```
